```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats

# Read the datasets
red_wine = pd.read_csv('winequality-red.csv', sep=';')
white_wine = pd.read_csv('winequality-white.csv', sep=';')

# Add wine type column
red_wine['wine_type'] = 'red'
white_wine['wine_type'] = 'white'

# Combine datasets
wines = pd.concat([red_wine, white_wine], axis=0)

# Set up the plotting style
plt.style.use('default')
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = [12, 8]

# Create a figure with multiple subplots
plt.figure(figsize=(20, 15))
```

```python
# 1. Distribution of Wine Quality
plt.subplot(2, 2, 1)
sns.boxplot(x='wine_type', y='quality', data=wines)
plt.title('Wine Quality Distribution by Type')
plt.xlabel('Wine Type')
plt.ylabel('Quality Score')
```

```python
# 2. Correlation with Alcohol Content
plt.subplot(2, 2, 2)
sns.scatterplot(data=wines, x='alcohol', y='quality', hue='wine_type', alpha=0.5)
plt.title('Quality vs Alcohol Content')
plt.xlabel('Alcohol (%)')
plt.ylabel('Quality Score')
```

```python
# Plot distributions of important features for red wine
plt.subplot(2, 3, 1)
sns.boxplot(x='quality', y='alcohol', data=red_wine)
plt.title('Alcohol Content vs Quality (Red Wine)')

plt.subplot(2, 3, 2)
sns.boxplot(x='quality', y='volatile acidity', data=red_wine)
plt.title('Volatile Acidity vs Quality (Red Wine)')

plt.subplot(2, 3, 3)
sns.boxplot(x='quality', y='sulphates', data=red_wine)
plt.title('Sulphates vs Quality (Red Wine)')
```

```python
# Plot distributions of important features for white wine
plt.subplot(2, 3, 4)
sns.boxplot(x='quality', y='alcohol', data=white_wine)
plt.title('Alcohol Content vs Quality (White Wine)')

plt.subplot(2, 3, 5)
sns.boxplot(x='quality', y='density', data=white_wine)
plt.title('Density vs Quality (White Wine)')

plt.subplot(2, 3, 6)
sns.boxplot(x='quality', y='residual sugar', data=white_wine)
plt.title('Residual Sugar vs Quality (White Wine)')

plt.tight_layout()
```
